```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) # comma between fig and ax allows us to unpack the tuple that subplots returns. fig is the figure object, ax is the axes object. 
# figsize is a tuple that specifies the width (5 in) and height (3 in) of the figure in inches.
ax.scatter(x, y) # scatter plot of x and y; since it's a scatter plot, it will plot each point somewhere on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.bar(x, y) # bar plot of x and y; since it's a bar plot, it will plot each point as a bar on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y) # line plot of x and y; since it's a line plot, it will plot each point as a line on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.hist(y) # histogram of y; since it's a histogram, we only need to input the y values, and it will plot the frequency of each value in y as a bar on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.hist(y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y) # line plot of x and y; since it's a line plot, it will plot each point as a line on the axes.
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```
