```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613) # set random seed for reproducibility
x = np.arange(50) # create an array of integers from 0 to 49
y = np.random.randint(0, 100, 50) # create an array of 50 random integers between 0 and 100
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) # comma between fig and ax allows us to unpack the tuple that subplots returns. fig is the figure object, ax is the axes object. 
# figsize is a tuple that specifies the width (5 in) and height (3 in) of the figure in inches.
ax.scatter(x, y) # scatter plot of x and y; since it's a scatter plot, it will plot each point somewhere on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.bar(x, y) # bar plot of x and y; since it's a bar plot, it will plot each point as a bar on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y) # line plot of x and y; since it's a line plot, it will plot each point as a line on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.hist(y) # histogram of y; since it's a histogram, we only need to input the y values, and it will plot the frequency of each value in y as a bar on the axes.
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.hist(y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y) # line plot of x and y; since it's a line plot, it will plot each point as a line on the axes.
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right',
        frameon = True # add a frame around the legend
        ) 
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right',
        frameon = True # add a frame around the legend
        , shadow = True
        ) 
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
        frameon = True, #add frame around the legend
        fontsize = 12, #change font size
        ncol = 2, #specify number of columns
        shadow = True )
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
        bbox_to_anchor =(1, 1)) # 0,0 means the upper left corner of the legend will be at the upper left corner of the axes. 
# 1,1 means the upper right corner of the legend will be at the upper right corner of the axes.
fig.show()

# if you don't include the bbox_to_anchor argument, the legend will be placed in the upper left corner of the axes by default.
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!",
      ha='center',    ##alignment of text
      color = 'red',  ##modify font colour
      size = 10)      ##modify font size
fig.show()
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) # set the x-axis limits to 0 and 10, and the y-axis limits to 0 and 10
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) # set the x-axis limits to 0 and 10, and the y-axis limits to 0 and 10
ax.text(1, 5, ". Data: (1, 5)",     transform=ax.transData) # the coordinates of the text are in data coordinates, which means that (1, 5) is the point that is 1 on the x-axis and 5 on the y-axis.
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)",    transform=ax.transAxes), # the coordinates of the text are in axes coordinates, which means that (0.5, 0.1) is the point that is 50% of the way across the x-axis and 10% of the way up the y-axis.
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure) # the coordinates of the text are in figure coordinates, which means that (0.2, 0.2) is the point that is 20% of the way across the x-axis and 20% of the way up the y-axis of the entire figure.
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
          arrowprops=dict(facecolor='black'))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', 
            xy=(10, 95),
            xytext=(20, 94),
          arrowprops=dict(arrowstyle = "fancy",
                          color ="hotpink"))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator()) # this will remove the major ticks on the y-axis, which are the ticks that are labeled with numbers.
ax.xaxis.set_major_formatter(plt.NullFormatter()) # this will remove the major ticks on the x-axis, which are the ticks that are labeled with numbers.
```

```python
# We can also define a max number of ticks we want on our axis.
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# We can set tick marks at a particular interval using the MultipleLocator function.
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # this will set the major ticks on the x-axis at intervals of 5, so the ticks will be at 0, 5, and 10.
```

```python
# Rotate axis labels and horizontal alignment using xticks()
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right')
```
